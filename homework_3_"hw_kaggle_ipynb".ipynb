### Second way (Cross-Validation)

```python
from sklearn.compose import ColumnTransformer
```

Creating a preprocessor

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])
```

Making the pipline

```python
pipeline = make_pipeline(
    preprocessor,
    LogisticRegression(solver='liblinear')
)
```

Definding parameters for GridSearch

```python
param_grid = {
    'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]
}
```

GridSearch:

```python
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc', refit=True)
```

```python
X = data.drop(columns=['Churn'])
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
grid_search.fit(X_train, y_train)
```

```python
print("Лучший параметр C:", grid_search.best_params_)
print("ROC-AUC на тестовом наборе данных:", roc_auc_score(y_test, grid_search.predict_proba(X_test)[:, 1]))
```

**Лучший параметр C: {'logisticregression__C': 10}**
